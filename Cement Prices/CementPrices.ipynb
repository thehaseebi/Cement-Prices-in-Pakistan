```python
df.dropna(inplace=True)
```

```python
df.isnull().any()
```

```python
citi = ['Islamabad', 'Rawalpindi', 'Gujranwala', 'Sialkot', 'Lahore', 'Faisalabad', 'Sargodha', 
        'Multan', 'Bahawalpur', 'Karachi', 'Hyderabad', 'Sukkur', 'Larkana', 'Peshawar', 
        'Bannu', 'Quetta', 'Khuzdar', 'Pakistan']

df[citi] = df[citi].apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)
df[citi] = df[citi].astype(int)

df['Period End'] = pd.to_datetime(df['Period End'], format='%d-%b-%Y')
```

```python
df.info()
```

```python
sns.lineplot(x=df['Period End'], y=df['Pakistan'], data=df)
plt.xlabel('Date')
plt.ylabel('Prices')
plt.title('Prices of cement in Pakistan')
plt.xticks(rotation=45)
plt.show()
```
